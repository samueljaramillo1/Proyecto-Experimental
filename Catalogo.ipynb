### Funciones analisis error

```python
def error_func(x, s_a, s_b, s_c):
    sigma_a = np.sqrt(s_a)
    sigma_b = np.sqrt(s_b)
    sigma_c = np.sqrt(s_c)
    return np.sqrt((sigma_a*x**2)**2 + (x*sigma_b)**2 + sigma_c**2)


def error_vel(velocidades, sigmas_dlambda, lambdas_i):
    c = 300000000
    return (c/lambdas_i)*sigmas_dlambda

def error_vel_promedio(N, sigmas: list):
    suma = 0
    for i in range(len(sigmas)):
        suma += sigmas[i]**2
    return np.sqrt(suma)/N

def limpia(lista_diferencia, lista_longitudes, lista_pixeles, min, max):
    a = lista_diferencia*300000000/np.array(lista_longitudes)
    b = np.array(lista_longitudes)
    c = np.array(lista_pixeles)

    filtro_1 = a < max
    a = a[filtro_1]
    lista_diferencia = lista_diferencia[filtro_1]
    b = b[filtro_1]
    c = c[filtro_1]

    filtro_2 = a > min
    a = a[filtro_2]
    lista_diferencia = lista_diferencia[filtro_2]
    b = b[filtro_2]
    c = c[filtro_2]

    return lista_diferencia, b, c
```
